```python
# Laden aller nötigen Pakete
import dash
from dash import dcc, html, dash_table, Input, Output
import plotly.graph_objs as go
import pandas as pd
import yfinance as yf
import numpy as np
import statsmodels.api as sm
from datetime import datetime
from bs4 import BeautifulSoup
import requests
```

```python
# Matrikelnummer als Seed für die .sample()-Methode
matrikelnummer = 105310

# Abrufen der Liste der Unternehmen des S&P 100
sp100_tickers = pd.read_html('https://en.wikipedia.org/wiki/S%26P_100')[2]
sp100_tickers = sp100_tickers[['Symbol', 'Name']]
sp100_tickers = sp100_tickers[~sp100_tickers['Symbol'].isin(['BRK.A', 'BRK.B'])]

# Auswahl von 24 Unternehmen ohne Zurücklegen (AAPL wird später hinzugefügt)
personalized_set = sp100_tickers[sp100_tickers['Symbol'] != 'AAPL'].sample(24, random_state=matrikelnummer, replace=False)
aapl_row = sp100_tickers[sp100_tickers['Symbol'] == 'AAPL']
personalized_set = pd.concat([personalized_set, aapl_row], ignore_index=True)
tickers = personalized_set['Symbol'].tolist()
```

```python
# Setzen des Zeitraums
start_date = '2021-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')

# Laden der Daten für ein Beispielunternehmen und den S&P 500 Index
ticker_symbol = 'AAPL'  # Ersetzen Sie dies durch den gewünschten Unternehmens-Ticker
sp500_ticker = '^GSPC'  # Ticker für den S&P 500 Index

# Abrufen der Daten
data_company = yf.download(ticker_symbol, start=start_date, end=end_date)['Adj Close']
data_sp500 = yf.download(sp500_ticker, start=start_date, end=end_date)['Adj Close']

# Berechnung der diskreten Renditen
data_company_discrete = data_company.pct_change()
data_sp500_discrete = data_sp500.pct_change()

# Berechnung der log-Renditen
data_company_log = np.log(data_company / data_company.shift(1))
data_sp500_log = np.log(data_sp500 / data_sp500.shift(1))

# Zusammenfassung in einem DataFrame
result_df = pd.DataFrame({
    'Company_Discrete_Returns': data_company_discrete,
    'S&P500_Discrete_Returns': data_sp500_discrete,
    'Company_Log_Returns': data_company_log,
    'S&P500_Log_Returns': data_sp500_log
})

# Ausgabe der ersten Zeilen der Daten
print(result_df.head())

# Erläuterung zur Verwendung des S&P 500 anstelle des S&P 100:
# "Die Bevorzugung des S&P 500 resultiert aus seiner umfangreicheren Abdeckung der US-Wirtschaft. 
# Er umfasst 500 Unternehmen, im Gegensatz zum S&P 100, der nur 100 Unternehmen beinhaltet. 
# Dadurch ermöglicht der S&P 500 eine breitere und vielfältigere Marktbetrachtung."
```

```python
# dataframe für Tab 2


# Start- und Enddatum für die Datenerhebung
start_date = '2021-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')

# Leerer DataFrame für die Finanzdaten
finanzdaten = pd.DataFrame()

# Laden der historischen Finanzdaten für jeden Ticker
for ticker in tickers:
    # Lade die historischen Daten für den Ticker
    data = yf.download(ticker, start=start_date, end=end_date)

    # Berechne tägliche Renditen
    data['Daily Return'] = data['Adj Close'].pct_change()

    # Füge die Ticker-Bezeichnung hinzu
    data['Ticker'] = ticker

    # Füge die Daten zum DataFrame hinzu
    finanzdaten = pd.concat([finanzdaten, data])
finanzdaten.reset_index(inplace=True)
```

```python
# dataframe für Tab 1
api_key = '4539f4af903e09b2e9eb3a1ac674883d'

# Leerer DataFrame für die gesammelten Daten
unternehmensdaten = pd.DataFrame()

for ticker in tickers:
    url = f'https://financialmodelingprep.com/api/v3/profile/{ticker}?apikey={api_key}'
    try:
        response = requests.get(url)
        data = response.json()[0]
    except Exception:
        data = {}
    unternehmensdaten = unternehmensdaten.append({
        'Ticker': ticker,
        'Name': personalized_set[personalized_set['Symbol'] == ticker]['Name'].values[0],
        'Mitarbeiter': data.get('fullTimeEmployees'),
        'Sektor': data.get('sector'),
        'Marktkapitalisierung': data.get('mktCap'),
        'Homepage': data.get('website')
    }, ignore_index=True)
```

```python
# Erstellen der Dash-App

# Funktion zur Berechnung von Beta und Risiko
def calculate_beta_and_risk(stock_returns, market_returns):
    X = sm.add_constant(market_returns)
    model = sm.OLS(stock_returns, X).fit()
    beta = model.params[1]
    residuals = model.resid
    systematic_risk = beta ** 2 * market_returns.var()
    idiosyncratic_risk = residuals.var()
    return model.params[0], beta, systematic_risk, idiosyncratic_risk

# Funktion zur Extraktion der Logo-URLs (Ich habe es leider nicht geschafft alle Logos mithilfe von yahoo und Wikipedia zu extrahieren. Ein paar sind allerdings da.)
def extract_logo_url(company_name):
    try:
        base_url = f"https://en.wikipedia.org/wiki/{company_name.replace(' ', '_')}"
        response = requests.get(base_url)
        if response.status_code != 200:
            return None
        soup = BeautifulSoup(response.content, "html.parser")
        infobox = soup.find("table", class_="infobox")
        if not infobox:
            return None
        infobox_image = infobox.find("img")
        if not infobox_image:
            return None
        return "https:" + infobox_image['src']
    except Exception:
        return None

def extract_logo_url_yahoo_finance(ticker):
    try:
        base_url = f"https://finance.yahoo.com/quote/{ticker}"
        response = requests.get(base_url)
        if response.status_code != 200:
            return None
        soup = BeautifulSoup(response.content, "html.parser")
        infobox = soup.find("div", {"id": "quote-header-info"})
        if not infobox:
            return None
        infobox_image = infobox.find("img", {"alt": "Logo"})
        if not infobox_image:
            return None
        return infobox_image['src']
    except Exception:
        return None

# Extrahieren Sie die Logos von Wikipedia
unternehmensdaten['Logo'] = unternehmensdaten['Name'].apply(extract_logo_url)

def update_logos(unternehmensdaten):
    for index, row in unternehmensdaten.iterrows():
        if not row['Logo']:
            logo_url = extract_logo_url_yahoo_finance(row['Ticker'])
            if logo_url:
                unternehmensdaten.at[index, 'Logo'] = logo_url

update_logos(unternehmensdaten)

unternehmensdaten['Logo'] = unternehmensdaten['Logo'].apply(lambda x: f"![Logo]({x})" if x else "")

app = dash.Dash(__name__)

company_dropdown = dcc.Dropdown(
    id='company-dropdown',
    options=[{'label': ticker, 'value': ticker} for ticker in tickers],
    value=tickers[0],
    clearable=False
)

company_table = dash_table.DataTable(
    id='company-data-table',
    columns=[{"name": i, "id": i, "presentation": "markdown" if i == "Logo" else None} for i in unternehmensdaten.columns],
    data=[],
    style_table={'overflowX': 'auto'},
    style_cell={
        'height': 'auto',
        'minWidth': '100px', 'width': '100px', 'maxWidth': '200px',
        'whiteSpace': 'normal'
    },
    markdown_options={"html": True}
)

toggle_lines = dcc.Checklist(
    id='toggle-lines',
    options=[
        {'label': 'Zeige 10% und 90% Quantile', 'value': 'quantiles'},
        {'label': 'Zeige +/- eine Standardabweichung', 'value': 'std_dev'}
    ],
    value=[]
)

app.layout = html.Div([
    dcc.Tabs([
        dcc.Tab(label='Gesamtübersicht', children=[
            html.Div([
                dcc.Graph(
                    id='sp500-graph',
                    figure={
                        'data': [
                            go.Scatter(
                                x=data_sp500.index,
                                y=data_sp500,
                                mode='lines'
                            )
                        ],
                        'layout': go.Layout(
                            title='Verlauf des S&P 500 Index über die Zeit',
                            xaxis={
                                'title': 'Datum',
                                'rangeslider': {'visible': True}
                            },
                            yaxis={'title': 'Indexwert'}
                        )
                    }
                ),
            ])
        ]),
        dcc.Tab(label='Unternehmensanalyse', children=[
            html.Div([
                company_dropdown,
                html.Div(id='selected-company-info'),
                company_table,
                toggle_lines,
                dcc.Graph(id='company-graph')
            ])
        ])
    ])
])

@app.callback(
    Output('company-data-table', 'data'),
    [Input('company-dropdown', 'value')]
)
def update_table(selected_ticker):
    selected_data = unternehmensdaten[unternehmensdaten['Ticker'] == selected_ticker]
    return selected_data.to_dict('records')

@app.callback(
    Output('company-graph', 'figure'),
    [Input('company-dropdown', 'value'), Input('toggle-lines', 'value')]
)
def update_graph(selected_ticker, toggle_values):
    if not selected_ticker:
        return go.Figure()
    company_data = finanzdaten[finanzdaten['Ticker'] == selected_ticker]
    if company_data.empty:
        return go.Figure()
    daily_returns = company_data['Daily Return'].dropna()
    sp500_data = data_sp500.loc[company_data['Date']] if 'Date' in company_data else data_sp500.loc[company_data.index]
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=company_data['Date'] if 'Date' in company_data else company_data.index, y=company_data['Adj Close'], mode='lines', name='Adjusted Close'))
    fig.add_trace(go.Scatter(x=sp500_data.index, y=sp500_data, mode='lines', name='S&P 500', yaxis='y2'))
    fig.update_layout(
        title=f'Adjusted Close Preis für {selected_ticker} und S&P 500',
        yaxis=dict(title='Company Adj Close'),
        yaxis2=dict(title='S&P 500', overlaying='y', side='right'),
        xaxis=dict(title='Datum')
    )
    if 'quantiles' in toggle_values:
        quantiles = daily_returns.quantile([0.1, 0.9])
        fig.add_hline(y=quantiles[0.1], line_dash="dash", line_color="green", annotation_text="10% Quantil")
        fig.add_hline(y=quantiles[0.9], line_dash="dash", line_color="red", annotation_text="90% Quantil")
    if 'std_dev' in toggle_values:
        mean = daily_returns.mean()
        std = daily_returns.std()
        fig.add_hline(y=mean + std, line_dash="dot", line_color="blue", annotation_text="+1 Std Dev")
        fig.add_hline(y=mean - std, line_dash="dot", line_color="blue", annotation_text="-1 Std Dev")
    return fig

if __name__ == '__main__':
    app.run_server(debug=True, port=8054)
```
